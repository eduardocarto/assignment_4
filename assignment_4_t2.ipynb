**This Jupyter Notebook file is very simple**. **First**, a *shapefile* is loaded using ***GeoPandas*** and then visualized.

## Step 1: Loading the Shapefile using GeoPandas Library

In this step, we:

- Import the `geopandas` library.
- Define the path to the shapefile stored locally.
- Use the `read_file()` function to load the shapefile into a **GeoDataFrame (`gdf`)**.

The GeoDataFrame allows us to work with both attribute data (like a DataFrame) and geometry (points, lines, polygons).

```python
import geopandas as gpd

FILE = "shapefile/loja_zamora.shp"

gdf = gpd.read_file(FILE)
```

## Step 2: Visualizing the Shapefile on an Interactive Map

Here, we use the `explore()` method from GeoPandas to display the shapefile on an interactive map.

- Internally, this function uses **Folium** and **Leaflet.js** to create a web-based map.
- You can zoom in/out, pan, and interact with the features.
- Make sure the required packages (`folium`, `matplotlib`, `mapclassify`) are installed.

> This is useful for quickly validating the spatial distribution of features.

```python
gdf.explore()
```
